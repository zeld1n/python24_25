##1.Immobili
Definire la classe Immobile con i seguenti attributi

*   codice
*   metri quadrati
*   tipologia che può assumere solo i valori : monolocale, bilocale, trilocale, quadrilocale
*   indirizzo
*   città
*   prezzo
*   venduto (inizialmente uguale a false)


Crea un oggetto, assegna dei valori e visualizzali utilizzando il metodo print(oggetto)


```python
class Immobile:
    def __init__(self, codice, metri_quadri, tipologia, indirizzo, citta, prezzo):
        self.codice = codice
        self.metri_quadri = metri_quadri
        self.tipologia = tipologia
        self.indirizzo = indirizzo
        self.citta = citta
        self.prezzo = prezzo
        self.venduto = False

    def __str__(self):
        return (f"codice: {self.codice}, mq: {self.metri_quadri}, tipologia: {self.tipologia}, "
                f"indirizzo: {self.indirizzo}, citta: {self.citta}, prezzo: {self.prezzo}, "
                f"venduto: {self.venduto}")

immobile1 = Immobile("123", 100, "bilocale", "via Napoli, 4", "Magenta", 1200)
print(immobile1)

n = int(input("Quanti immobili vuoi inserire? "))
immobili = []
immobili.append(immobile1)

for i in range(n):
    codice = input(f"Inserisci il codice dell'immobile {i+1}: ")
    for immobileLista in immobili:
      cod=immobileLista.codice
      if(cod==codice):
        codice = input(f"Inserisci il codice dell'immobile {i+1}: ")

    mq = int(input("inserisci i mq: "))
    tipo = input("inserisci la tipologia (monolocale, bilocale, trilocale, quadrilocale): ")
    while(not(tipo=="monolocale" or tipo=="bilocale" or tipo=="trilocale" or tipo=="quadrilocale")):
       tipo = input("inserisci la tipologia (monolocale, bilocale, trilocale, quadrilocale): ")
    indirizzo = input("inserisci l'indirizzo: ")
    citta = input("inserisci la città: ")
    prezzo = int(input("inserisci il prezzo: "))
    immobili.append(Immobile(codice, mq, tipo, indirizzo, citta, prezzo))

for immobile in immobili:
    print(immobile)


class Agenzia:
    def __init__(self):
        self.immobili = []

    def aggiungi_immobile(self, immobile):
       for imm in self.immobili:
            if imm.codice == immobile.codice:
                print(f"Immobile con questo codice {immobile.codice} già esiste")
                return
       self.immobili.append(immobile)

    def vendi_immobile(self, codice):
        for immobile in self.immobili:
            if immobile.codice == codice:
                immobile.venduto = True
                print(f"L'immobile con codice {codice} è stato venduto.")
                return
        print(f"Immobile con codice {codice} non trovato.")

    def aggiorna_prezzo(self, codice, nuovo_prezzo):
        for immobile in self.immobili:
            if immobile.codice == codice:
                immobile.prezzo = nuovo_prezzo
                print(f"Prezzo dell'immobile con codice {codice} aggiornato a {nuovo_prezzo}€.")
                return
        print(f"Immobile con codice {codice} non trovato.")

    def visualizza_per_tipologia(self, tipologia):
        print(f"Immobili di tipologia {tipologia}:")
        for immobile in self.immobili:
            if immobile.tipologia == tipologia:
                print(immobile)

    def cerca_immobile(self, codice):
        for immobile in self.immobili:
            if immobile.codice == codice:
                print(immobile)
                return
        print(f"Immobile con codice {codice} non trovato.")

    def visualizza_tutti(self):
        print("Tutti gli immobili:")
        for immobile in self.immobili:
            print(immobile)

    def totale_prezzi_venduti(self):
        totale = 0
        for immobile in self.immobili:
          if immobile.venduto:
              totale += immobile.prezzo
        print(f"Totale prezzi degli immobili venduti: {totale}")

    def prezzo_medio_tipologia(self, tipologia):
      totale_prezzo = 0
      conteggio = 0
      for immobile in self.immobili:
          if immobile.tipologia == tipologia:
              totale_prezzo += immobile.prezzo
              conteggio += 1
      if conteggio > 0:
          prezzo_medio = totale_prezzo / conteggio
          print(f"Prezzo medio degli immobili di tipologia {tipologia}: {prezzo_medio:.2f}€")
      else:
          print(f"Nessun immobile di tipologia {tipologia} trovato.")


immobile1 = Immobile("0012", 50, "bilocale", "via Roma 1", "Milano", 150000)
immobile2 = Immobile("00233", 70, "trilocale", "via Milano 20", "Torino", 200000)
immobile3 = Immobile("003444", 40, "monolocale", "via Napoli 15", "Roma", 100000)
agenzia = Agenzia()
for imob in immobili:
  agenzia.aggiungi_immobile(imob)
"""MENU"""
while(True):
  scelta=int(input("""
  1.Aggiunta di un immobile
  2.Vendita di un immobile
  3.Aggiorna prezzo
  4.Visualizza immobili appartenenti ad una tipologia
  5.Cerca un immobile
  6.Visualizza tutti gli immobili
  7.Visualizza il totale dei prezzi di tutti gli immobili venduti
  8.Visualizza il prezzo medio di una tipologia fornita in input
  """))
  if(scelta==1):
    agenzia.aggiungi_immobile(immobile1)
    agenzia.aggiungi_immobile(immobile2)
    agenzia.aggiungi_immobile(immobile3)

  elif(scelta==2):
    vendi=input("Inserisci codice")
    agenzia.vendi_immobile(vendi)
  elif(scelta==3):
    codice=input("Inserisci codice: ")
    prezzo=int(input("Inserisci il prezzo: "))
    agenzia.aggiorna_prezzo(codice,prezzo)
  elif(scelta==4):
    tipo=input("Inserisci tipologia: ")
    agenzia.visualizza_per_tipologia(tipo)
  elif(scelta==5):
    codice=input("Inserisci codice: ")
    agenzia.cerca_immobile(codice)
  elif(scelta==6):
    agenzia.visualizza_tutti()
  elif(scelta==7):
    agenzia.totale_prezzi_venduti()
  elif(scelta==8):
    tipologia_input = input("Inserisci la tipologia di cui vuoi il prezzo medio: ")
    agenzia.prezzo_medio_tipologia(tipologia_input)
  elif(scelta==0):
    break




















```
